### Preprocessing
We used MinMaxScaler scale all the values in between 0 and 1

```python
def scale(xtrain, xtest, xval):
    scaler = MinMaxScaler().fit(xtrain)
    
    X_train_scaled = scaler.transform(xtrain)
    X_val_scaled = scaler.transform(xval)
    X_test_scaled = scaler.transform(xtest)
    
    return X_train_scaled, X_val_scaled, X_test_scaled
```

### Split training and testing

```python
def split (features):
    result = train_test_split(features, target, random_state=3000)
    return result
```

```python
def split_the_dataset(x1, x2):
        #split data into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(x1, x2, random_state=3000)
        
        return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = split_the_dataset(features, target)
X_train, X_val, y_train, y_val = split_the_dataset(X_train, y_train)
```

```python
X_train_scaled, X_val_scaled, X_test_scaled = scale(X_train, X_test, X_val)
```

```python
# model dictionary
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

estimators = {'Naive Bayes': GaussianNB(),
              'k Nearest Neighbors': KNeighborsClassifier(), 
              'Support Vector Machine': LinearSVC(max_iter=1000000),
             'Logistic Regression' : LogisticRegression(),
             'Decision Tree Regressor': DecisionTreeClassifier()}
```

### Feature Selection

```python
from sklearn.feature_selection import SelectFromModel

def selection(features, x_train, x_test, x_val, ytrain):
    select = SelectFromModel(DecisionTreeRegressor(random_state= 3000), threshold='median')
    #fit scaled and y train
    select.fit(x_train, ytrain)
    #transform previous scaled sets to only select 3 most important features in each
    X_train_selected = select.transform(x_train)
    X_val_selected = select.transform(x_val)
    X_test_selected = select.transform(x_test)
    
    #determine selected features
    selected_features = [feature for feature, status in zip(features, select.get_support()) if status == True]
    print('Selected features:')
    for feature in selected_features:
        print('\t' + feature)

    return X_train_selected, X_val_selected, X_test_selected
```

```python
# X_train_selected: Subset of training set. Has only selected features
# X_test_selected: Subset of testing set. Has only selected features
X_train_selected, X_val_selected, X_test_selected = selection(features, X_train_scaled, X_test_scaled, X_val_scaled, y_train)
```

```python
import numpy as np
import scipy.stats as stats
from scipy.stats import mannwhitneyu

# evaluate p value
def p_eval(p, name):
    if p < 0.05:
        print('Since the p-value of ', str(p),
              ' is less than 0.05, we reject the null hypothesis.', 'We have evidence that ',
              name + ' effects an individuals risk of CHD')
    else:
        print('Since the p-value of ', str(p),
              ' is more than 0.05, we cannot reject the null hypothesis.', 
              'We have no evidence that ', name + ' effects an individuals risk of CHD.')
```

### 3.5. Model Optimization

#### Indicate the best models:

```python
best_estimators = {
    'Support Vector Machine': LinearSVC(max_iter=1000000),
    'Logistic Regression' : LogisticRegression()
}
```

#### Rationale Behind Tuning
We chose the Support Vector Machine and the Logistic Regression because they had the best accuracy scores. The Support Vector Machine is simple, easy to train and scales well to very large datasets, so its high performance is expected. The Logistic Regression adds a layer of complexity as it uses the sigmoid function in case there is not a linear relationship between the features and it adds weights to the features which would be extremely useful for this dataset. 
 
In terms of tuning, we tune because:
* Our model needs to be accurate on the training and test set
* To prevent overfitting where the model is too complex and becomes too specific to our training set. 
* to prevent underfitting where the model is too simple in which it cannot encompass the entire dataset.
* to optimizes the generalizability of the algorithm
* to choose the parameter values that lead to the best predictions. 

#### Tuning

```python
from sklearn.model_selection import GridSearchCV
params = {'C' : [0.1, 1, 5, 10, 20, 30, 50]}

grid_params = {}
for name, estimator in best_estimators.items():
    grid = GridSearchCV(estimator, params, cv=5)
    grid.fit(X=X_train_selected, y=y_train)
    
    print(name)
    print('\tBest cross-validation score: ', grid.best_score_)
    print('\tBest parameters: ', grid.best_params_)
    
    grid_params[name] = grid.best_params_['C']
```

### 3.6. Model Testing

#### Testing the data with our best parameters and estimators

```python
for name, estimator in best_estimators.items():
    model = estimator.set_params(C = grid_params[name]).fit(X_train_selected, y=y_train)
    print(name, ' with tuned C value of ', grid_params[name], ":")
    print('\tPrediction accuracy on test data: ', model.score(X_val_selected, y_val))
    
```
